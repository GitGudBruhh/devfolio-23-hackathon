```python
def count_words(text):
    words = word_tokenize(text)
    return len(words)
df['count']=df['Subject'].apply(count_words)
df['count']
```

```python
%%time
def clean_str(string, reg = RegexpTokenizer(r'[a-z]+')):
    # Clean a string with RegexpTokenizer
    string = string.lower()
    tokens = reg.tokenize(string)
    return " ".join(tokens)

print('Before cleaning:')
df.head()
```

```python
print('After cleaning:')
df['Subject'] = df['Subject'].apply(lambda string: clean_str(string))
df.head()
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
def stemming (text):
    return ''.join([stemmer.stem(word) for word in text])
df['Subject']=df['Subject'].apply(stemming)
df.head()
```

```python
X = df.loc[:, 'Subject']
y = df.loc[:, 'Label_Number']

print(f"Shape of X: {X.shape}\nshape of y: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=11)
```

```python
print(f"Training Data Shape: {X_train.shape}\nTest Data Shape: {X_test.shape}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
cv.fit(X_train)
```
